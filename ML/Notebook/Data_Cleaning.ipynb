```python
columns_to_drop = ['saturated_fat_g', 'cholesterol_mg', 'dietary_fiber_g', 'sodium_mg', 'calories_from_fat', 'calcium_mg', 'iron_mg', 'magnesium_mg', 'potassium_mg', 'zinc_mg', 'phosphorus_mg', 'vitamin_a_iu_IU',
                   'niacin_equivalents_mg', 'vitamin_b6_mg', 'vitamin_c_mg', 'folate_mcg', 'thiamin_mg', 'riboflavin_mg', 'vitamin_e_iu_IU', 'vitamin_k_mcg', 'biotin_mcg', 'vitamin_b12_mcg', 'mono_fat_g',
                   'poly_fat_g', 'trans_fatty_acid_g', 'omega_3_fatty_acid_g', 'omega_6_fatty_acid_g']
df_merged2 = df_merged.drop(columns_to_drop, axis = 1)
```

Berdasarkan output dan diatas, dapat disimpulkan bahwa data null pada kolom 'prep' dan 'total 'berkorespondensi dengan data null pada kolom 'cook' yang menyebabkan ketiganya tidak dapat dicari nilainya. Jadi akan diisi dengan meannya

kolom 'prep' dan 'total' akan didrop.

```python
df_merged2['author'].fillna('Unknown', inplace=True)
df_merged2 = df_merged2.drop('prep', axis='columns')
df_merged2 = df_merged2.drop('total', axis='columns')
```

```python
def convert_to_timedelta(time_str):
    if pd.isna(time_str) or isinstance(time_str, float):
        return pd.NaT
    # Mengganti singkatan yang tidak dikenali dengan yang dikenali
    time_str = time_str.replace('mins', 'minutes')
    time_str = time_str.replace('hrs', 'hours')
    # Menggunakan pd.to_timedelta untuk berbagai format
    return pd.to_timedelta(time_str)

# Mengubah kolom 'waktu_persiapan' ke timedelta
df_merged2['cook'] = df_merged2['cook'].apply(convert_to_timedelta)

print("\nDataFrame setelah mengubah kolom 'cook' ke timedelta:")
print(df_merged2)
print("\nDtypes DataFrame:")
print(df_merged2.dtypes)
```

```python
df_merged2['cook'].fillna(df_merged2['cook'].mean(), inplace = True)
```
